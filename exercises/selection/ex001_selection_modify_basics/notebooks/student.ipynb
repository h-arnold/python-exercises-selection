## Goal
Welcome to **selection** — making decisions in code! In these exercises you will practise modifying existing programs that use `if` statements.

By the end of this notebook you will be able to:
- Change what a program checks using `==`, `!=`, `<`, and `>`
- Change what a program prints when a condition is true
- Use `if-else` to handle both true and false conditions
- Predict how changing a comparison affects the program's behaviour

## Comparison Operators
These are the four operators you will use:

| Operator | Meaning |
|----------|---------|
| `==` | equal to |
| `!=` | not equal to |
| `<` | less than |
| `>` | greater than |

## Introducing `else`

So far, each program has only done something when a condition was **True**. But what if you want to do one thing when the condition is **True** and a different thing when it is **False**? That is where `else` comes in!

```python
if condition:
    # do this when True
else:
    # do this when False
```

The exercises from here on already include an `else` branch. You will modify both what happens when the condition is True **and** what happens when it is False.
